##Function 1
$f1 = cos(x) + cos(y)$

```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection='3d')

x = np.linspace(0,2*np.pi)
y = np.linspace(0,2*np.pi)
z = (np.cos(x) + np.cos(y))

ax.plot3D(x,y,z)
ax.plot([np.pi],[np.pi],[-2],'bo')
```

Above is the three dimensional plot of the function. Here, the blue dot is p1 (the minimum).

```
def f1(x,y):
  return (np.cos(x) + np.cos(y))

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)
plt.contour(X,Y,Z, colors = 'black')
plt.plot([np.pi],[np.pi],'bo')
```

This is the contour plot of the function. The blue point is p1 (the minimum).

```
plt.contourf(X,Y,Z,20, cmap='RdGy')
plt.colorbar();
plt.plot([np.pi],[np.pi],'bo')
```

This is the density plot of the function. Here you can see that the red areas are valleys and the black areas are peaks, Thus, you can see that the darkest red middle area is p1 (the minimum). I have plotted a blue point so you can see exactly where it is.

#Function 2
$f2 = sin(x)+sin(y)$

```
fig = plt.figure()
ax = plt.axes(projection='3d')
x = np.linspace(0,np.pi)
y = np.linspace(0,np.pi)
z = (np.sin(x) + np.sin(y))

zmax = np.argmax(z)

ax.plot3D(x,y,z)
ax.plot([np.pi/2],[np.pi/2], [2], 'bo')
```

Here p2 (maximum) is the blue point.



```
def f2(x,y):
  return (np.sin(x) + np.sin(y))

X, Y = np.meshgrid(x, y)
Z = f2(X, Y)
plt.contour(X,Y,Z, colors = 'black')
plt.plot([np.pi/2],[np.pi/2], 'bo')
```

Here the maximum p2 is the blue point.

```
plt.contourf(X,Y,Z,20, cmap='RdGy')
plt.colorbar();
plt.plot([np.pi/2],[np.pi/2], 'bo')
```

Here I have marked the maximum p2 with a blue point.

#Function 3
$f3 = x^3 + y^3$

```
fig = plt.figure()
ax = plt.axes(projection='3d')
x = np.linspace(-5,5)
y = np.linspace(-5,5)
z = (x**3 + y**3)

ax.plot3D(x,y,z)
ax.plot([0],[0],[0], 'bo')
```

Here, p3 is located at the point of inflection and is notated by the blue point.

```
def f2(x,y):
  return (x**3 + y**3)

X, Y = np.meshgrid(x, y)
Z = f2(X, Y)
plt.contour(X,Y,Z, colors = 'black')
plt.plot([0],[0],[0], 'bo')
```

p3 in this graph is the blue point.

```
plt.contourf(X,Y,Z,20, cmap='RdGy')
plt.colorbar();
plt.plot([0],[0],[0], 'bo')
```

p3 in this graph is also the blue point.
